```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('fivethirtyeight')
plt.rc('figure', figsize=(5.0, 2.0))
```

```python
games=pd.read_csv('games.csv',sep=',', index_col=0)
```

```python
games
```

```python
aperture=dict()
with open('eco.txt','r') as f:
    for line in f:
        line=line.strip('\n')
        campi=line.split(':')
        lista_eco=[]
        if '-' in campi[0]:
            estremi_codici=campi[0].split('-')
            lettera=estremi_codici[0][0]
            inizio=estremi_codici[0][1:]
            fine=estremi_codici[1][1:]

            for x in range(int(inizio),int(fine)+1):
                if x < 10:
                    numero='0'+str(x)
                else:
                    numero=str(x)
                lista_eco.append(lettera+numero)
        else:
            lista_eco.append(campi[0])

        for x in lista_eco:
            aperture[x]=campi[1]
```

```python
aperture
```

```python
freq_aperture=pd.crosstab(index=games['opening_eco'],
                         columns=['Frequenza Assoluta'],
                         colnames=[''])
freq_aperture
```
